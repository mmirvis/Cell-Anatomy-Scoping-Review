```python
# from BioPython Entrez automatic search.ipynb
# takes in id_list from search function output, returns a dictionary with items 
# PubmedArticle and PubmedBookArticle, which are lists of info about each article entry, including metadata, title, etc

def fetch_details(id_list):
#     ids = ','.join(id_list)
    Entrez.email = 'mariya.mirvis@ucsf.edu'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=id_list)
    results = Entrez.read(handle)
    return results
```

```python
# perform Entrez fetch on list of PMIDs from 1st round of screening

stage1_fetch = fetch_details(coll_031523['PMID'].values)
```

```python
# pull abstracts and sort into categories, save results in a dataframe
# read abstract, type E, C, or I 
# "Include" criteria: whole-cell imaging including multiple organelles
# comments: very brief, reason for decision, if "C" - what to look for; make note if prokaryotic system
# must say "Y" to stop before moving on, otherwise it gets stuck in a loop and you may lose progress

stage2_sort = pd.DataFrame(columns = ['Index', 'PMID', 'Exclude (E)/Check Full Text (C)/Include (I)', 'Comments'])
stage2_sort
index = 1 -1 # update first number after stopping and returning to where you left off, should start at PMID 36418108, always subtract by 1

for item in range(len(coll_031523['PMID'])):
    title = coll_031523['Title'][index]
    print(coll_031523['PMID'][index],"Title: ", title)

    # some Pubmed entries lack an abstract, this would create an error otherwise
    if 'Abstract' in stage1_fetch['PubmedArticle'][index]['MedlineCitation']['Article']:
        abstract = stage1_fetch['PubmedArticle'][index]['MedlineCitation']['Article']['Abstract']['AbstractText']
        print("Abstract: ", abstract)

    else:
        print("No abstract available, indicate Check Full Text (C)")

    sort_ECI = input('Exclude (E), Check Full Text (C), or Include (I)?')
    while sort_ECI not in ['E', 'C', 'I']:
        sort_ECI = print('whoops! invalid answer')
        sort_ECI = input('Exclude (E), Check Full Text (C), or Include (I)?')

    cmmts = input('Comments: ')
#     comments.append(cmmts)

#     print(stage2_sort)
#     print(index,item,sort_ECI,cmmts)
    stage2_sort.loc[len(stage2_sort)] = [index,coll_031523['PMID'][index],sort_ECI,cmmts]
    
    stop = input('Stop? (Y/N)')
    if stop == 'Y':
        break
        
    index+=1
```
